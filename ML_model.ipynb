```python
# --- Clean Naukri dataset ---
df_naukri_clean = df_naukri[['jobtitle', 'company', 'joblocation_address', 'skills', 'industry', 'experience', 'payrate']].copy()
df_naukri_clean.rename(columns={
    'jobtitle': 'job_title',
    'company': 'company_name',
    'joblocation_address': 'location',
    'skills': 'description',
    'industry': 'category',
    'experience': 'experience',
    'payrate': 'salary'
}, inplace=True)

print("‚úÖ Naukri cleaned shape:", df_naukri_clean.shape)


# --- Clean Indeed dataset ---
df_indeed_clean = df_indeed[['job_title', 'company_name', 'city', 'job_description', 'category']].copy()
df_indeed_clean.rename(columns={
    'city': 'location',
    'job_description': 'description'
}, inplace=True)

print("‚úÖ Indeed cleaned shape:", df_indeed_clean.shape)


# --- Merge both ---
combined_df = pd.concat([df_naukri_clean, df_indeed_clean], ignore_index=True)

# Drop duplicates and rows without description/title
combined_df.drop_duplicates(subset=['job_title', 'company_name', 'description'], inplace=True)
combined_df.dropna(subset=['job_title', 'description'], inplace=True)

print("‚úÖ Combined dataset shape:", combined_df.shape)
combined_df.head()
```

```python
# Add a new column to store suitable disability types
combined_df['suitable_for_disability'] = ''

# Loop through each disability type and its keywords
for _, row in df_disability.iterrows():
    disability = row['disability_type']
    keywords = [kw.strip().lower() for kw in row['keywords'].split('|')]
    
    # For each keyword, check if it appears in job title or description
    mask = combined_df['job_title'].str.lower().str.contains('|'.join(keywords), na=False) | \
           combined_df['description'].str.lower().str.contains('|'.join(keywords), na=False)
    
    # Assign disability tag where match found
    combined_df.loc[mask, 'suitable_for_disability'] += disability + '|'

# Clean up column (remove trailing "|")
combined_df['suitable_for_disability'] = combined_df['suitable_for_disability'].str.strip('|')

print("‚úÖ Disability suitability mapping complete!")
combined_df[['job_title', 'suitable_for_disability']].head(15)
```

```python
import pandas as pd

# Just in case your variable name differs
df = combined_df.copy()

# Fill missing values
df['job_title'] = df['job_title'].fillna('')
df['description'] = df['description'].fillna('')

# Combine into one text field
df['text'] = df['job_title'] + ' ' + df['description']

print("‚úÖ Combined text column created!")
df[['job_title', 'text']].head(10)
```

```python
# Keep only relevant columns for text modeling
keep_cols = ['job_title', 'company_name', 'location', 'description', 'category', 'suitable_for_disability']
df = df[keep_cols].copy()
```

```python
# Recreate the description length column using correct name
df['desc_length'] = df['description'].fillna('').apply(len)
print(df['desc_length'].describe())
```

```python
# Keep only rows where description length is meaningful
df = df[df['desc_length'] > 50].copy()

# Combine text columns
df['text'] = (
    df['job_title'].fillna('') + " " +
    df['description'].fillna('') + " " +
    df['category'].fillna('')
)

# Clean text
import re

def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', str(text))  # remove punctuation
    text = text.lower()                                # lowercase
    text = re.sub(r'\s+', ' ', text).strip()           # remove extra spaces
    return text

df['text'] = df['text'].apply(clean_text)

print("‚úÖ Filtered dataset shape:", df.shape)
print("Example text:\n", df['text'].iloc[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer



tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1,2), max_features=10000)
tfidf_matrix = tfidf.fit_transform(df['text'])

print("‚úÖ TF-IDF matrix created successfully!")
print("Matrix shape:", tfidf_matrix.shape)

```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_jobs(user_input, top_n=5):
    """
    Recommend jobs based on user skills using TF-IDF + Cosine Similarity
    """
    # Transform user input into TF-IDF vector
    user_vec = tfidf.transform([user_input])

    # Compute cosine similarity between user input and job postings
    similarity = cosine_similarity(user_vec, tfidf_matrix).flatten()

    # Get indices of top N similar jobs
    top_indices = similarity.argsort()[-top_n:][::-1]

    # Select the matching jobs
    results = df.iloc[top_indices][['job_title', 'company_name', 'location', 'description', 'suitable_for_disability']]
    results['similarity_score'] = similarity[top_indices]

    return results.reset_index(drop=True)

print("‚úÖ Job recommendation function created successfully!")
```

```python
user_input = "Python, SQL, Data Analysis"
recommendations = recommend_jobs(user_input, top_n=5)

print("üîç Top job recommendations for:", user_input)
recommendations
```

```python
def recommend_jobs_inclusive(user_input, disability_type=None, top_n=5):
    user_vec = tfidf.transform([user_input])
    similarity = cosine_similarity(user_vec, tfidf_matrix).flatten()
    df['similarity_score'] = similarity
    sorted_df = df.sort_values(by='similarity_score', ascending=False)

    # --- Rule-based logic: disability-friendly job categories ---
    suitable_categories = {
        "hearing_impairment": ["IT", "Data", "Software", "Finance", "Analysis", "Admin", "Research"],
        "visual_impairment": ["HR", "Teaching", "Support", "Training", "Writing", "Recruitment"],
        "mobility_impairment": ["Remote", "Data Entry", "IT", "Finance", "Consulting", "Analysis"],
        "speech_impairment": ["IT", "Data", "Design", "Finance", "Research", "Admin"],
        "cognitive_disability": ["Manual", "Packaging", "Assembly", "Helper", "Support"]
    }

    # Apply text-based disability filtering + rule-based inclusivity
    if disability_type:
        # Start by selecting jobs that are explicitly marked suitable
        filtered_df = sorted_df[
            sorted_df['suitable_for_disability'].str.contains(disability_type, case=False, na=False)
            | sorted_df['suitable_for_disability'].eq('')
        ]

        # Add rule-based jobs for this disability (by category or job title)
        if disability_type in suitable_categories:
            rules = suitable_categories[disability_type]
            rule_based_jobs = sorted_df[
                sorted_df['job_title'].str.contains('|'.join(rules), case=False, na=False)
                | sorted_df['category'].str.contains('|'.join(rules), case=False, na=False)
            ]
            # Combine both sets (explicitly suitable + logically suitable)
            filtered_df = pd.concat([filtered_df, rule_based_jobs]).drop_duplicates()
    else:
        filtered_df = sorted_df

    # --- Return top N recommendations ---
    recommendations = filtered_df.head(top_n)[
        ['job_title', 'company_name', 'location', 'category', 'suitable_for_disability', 'similarity_score']
    ].reset_index(drop=True)

    return recommendations
```

```python
user_input = "Python, SQL, Data Analysis"
user_disability = "visual_impairment"  # can change to 'hearing_impairment', etc.

recommendations = recommend_jobs_inclusive(user_input, disability_type=user_disability, top_n=5)

print(f"üîç Top job recommendations for: '{user_input}' (Disability: {user_disability})")
recommendations
```
